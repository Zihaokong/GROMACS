```python
import csv
import numpy as np
import matplotlib.pyplot as plt

data = np.loadtxt(open("Gromacs Test Result.csv", "rb"), delimiter=",", skiprows=1).T


# Create some mock data
t = data[0]
data1 = data[1]
data2 = data[2]
 
fig, ax1 = plt.subplots()
fig.set_size_inches(10.5, 8.5)
ax1.set_xlabel('Num of MPI Processes(s)',fontsize=15)
ax1.set_ylabel('Performance (Ns/Day)', fontsize=15,color='tab:green')
l1 =ax1.plot(t, data1,color='tab:green',marker='o',label="Performance")
ax1.tick_params(axis='y', labelcolor='tab:green',size=10,labelsize=15)

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis
ax1.legend(loc=1,fontsize=15)
color = 'tab:blue'
ax2.set_ylabel("Wall time (s)", color=color,fontsize=15)  # we already handled the x-label with ax1
l2 = ax2.plot(t, data2, color=color,marker='o',label="Wall Time")
ax2.legend(loc=7,fontsize=15)
ax2.tick_params(axis='y', labelcolor=color,labelsize=15)
fig.tight_layout()  # otherwise the right y-label is slightly clipped




plt.show()
```
